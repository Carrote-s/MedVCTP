```python
from transformers import AutoProcessor, AutoModelForImageTextToText
from PIL import Image
import torch
from transformers import AutoProcessor, AutoModelForImageTextToText
from PIL import Image
import requests
import torch

model_id = "google/medgemma-4b-it"

model = AutoModelForImageTextToText.from_pretrained(
    model_id,
    torch_dtype=torch.bfloat16,
    device_map="auto",
)

processor = AutoProcessor.from_pretrained(model_id)
```

```python
import json
with open('entriesMainClosed', "r") as f:
    entriesClosed = json.load(f)
```

```python
def decode(inputImg):
    messages = [
        {
            "role": "system",
            "content": [{"type": "text", "text": "You are an expert radiologist."}]
        },
        {
            "role": "user",
            "content": [
                {"type": "text", "text": """Describe this medical image with a focus on all key observations, abnormal findings, and high-level features. Include the following:
    Imaging parameters: modality (MRI, X-ray, CT, etc.), plane (axial, sagittal, coronal), contrast administration (enhanced or not), exposure or image quality.
    Structures visualized: all relevant organs, vessels, and anatomical regions visible in the image (e.g., ventricles, cerebellum, arteries, colon, lung parenchyma).
    Abnormalities: any lesions, masses, fluid collections, air, fractures, atrophy, obstruction, enhancement patterns, or other deviations from normal. 
    Describe location (cortical, subcortical, lobe, organ), size (relative or absolute if measurable), shape, number, and relationship to adjacent structures.
    Structural relationships: midline symmetry, displacement, compression, mass effect, or shifts.
    Overall assessment: note if the image appears grossly normal or abnormal without giving a final diagnosis.
    Provide enough detail that a clinician or AI model could answer yes/no questions about lesion presence, organ visualization, structural abnormalities, or imaging features. 
    Avoid repeating disclaimers, unrelated commentary, or giving a diagnosis.
    Provide all requested details, but be concise and avoid unnecessary repetition to keep the output short enough for processing.
    """},
                {"type": "image", "image": inputImg}
            ]
        }
    ]
    
    inputs = processor.apply_chat_template(
        messages, add_generation_prompt=True, tokenize=True,
        return_dict=True, return_tensors="pt"
    ).to(model.device, dtype=torch.bfloat16)
    
    input_len = inputs["input_ids"].shape[-1]
    with torch.inference_mode():
        generation = model.generate(**inputs, max_new_tokens=700)
        generation = generation[0][input_len:]
    
    decoded = processor.decode(generation, skip_special_tokens=True)
    return decoded
```
